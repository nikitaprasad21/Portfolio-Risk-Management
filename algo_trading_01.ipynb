## Moving Average / Rolling Mean

A moving average (MA) is a statistical calculation used in finance and other fields to analyze data over a certain period, smooth out fluctuations, and identify trends. It is a commonly used technical analysis indicator in the world of trading and investing. Moving averages are particularly useful for identifying trend directions and potential support or resistance levels.

There are several types of moving averages, with two of the most common ones being:

Simple Moving Average (SMA):
The Simple Moving Average calculates the average price of an asset over a specific number of data points (typically closing prices) for a given time period. It assigns equal weight to all data points within that period. For example, a 50-day SMA would sum up the closing prices of the last 50 days and divide the total by 50.

Exponential Moving Average (EMA):
The Exponential Moving Average gives more weight to recent data points, making it more responsive to current price changes. This type of moving average reduces lag and is often preferred for short-term analysis. EMA places more importance on the most recent prices by applying a weighting factor.

When to use each type of moving average:

-- SMA:

Use SMA when you want a straightforward and less reactive representation of price data.
It is suitable for identifying longer-term trends and support/resistance levels.
Traders and investors often use longer SMAs (e.g., 200-day SMA) to analyze overall market trends and long-term investment decisions.

-- EMA:

EMA is preferred for short-term analysis and trading due to its responsiveness to recent price changes.
It's valuable for identifying short-term trends and potential entry/exit points.
Day traders or swing traders may use shorter EMAs (e.g., 10-day or 20-day) for quick decision-making.


In addition to SMA and EMA, there are other variations like the Weighted Moving Average (WMA) and the Adaptive Moving Average (AMA), each with its own unique characteristics and use cases. The choice of which moving average to use depends on your trading or investment strategy and the specific timeframes you are interested in.

It's essential to note that moving averages are just one part of technical analysis, and traders often use them in combination with other indicators and analysis techniques to make informed decisions. The choice of moving average and the parameters (e.g., time periods) should align with your trading strategy and goals

```python
def calculate_and_plot_ma_ewma_for_all_companies(close_data):
    # Define a list of company symbols
    company_symbols = ["AMZN",	"GOOGL", "META", "MSFT", "NFLX"]

    for company_symbol in company_symbols:
        # Create a new DataFrame for the specified company's data
        company_ma = close_data[company_symbol].copy().to_frame()
        
        # Calculate and add the 50-day Simple Moving Average (SMA50)
        company_ma['SMA50'] = company_ma[company_symbol].rolling(window=50).mean()
        
        # Calculate and add the 200-day Simple Moving Average (SMA200)
        company_ma['SMA200'] = company_ma[company_symbol].rolling(window=200).mean()
        
        # Calculate and add the 30-day Exponential Moving Average (EWMA30)
        company_ma['EWMA30'] = company_ma[company_symbol].ewm(span=30).mean()
        
        # Plot the data for the specific company
        company_ma.plot(figsize=(20, 8), fontsize=15)
        plt.title(f"{company_symbol} Moving Averages", fontsize=15)
        plt.legend(loc="upper left", fontsize=15)
        plt.show()

# Example usage:
# Assuming you have a DataFrame 'close' with data for multiple companies
# you can call the function to analyze and plot moving averages for all companies like this:
# calculate_and_plot_ma_ewma_for_all_companies(close)
```

## Cumulative Returns:
Cumulative returns represent the total return of an investment over a given period, showing the growth or decline in value over time. 

The choice between cumsum and cumprod depends on what you want to represent in the cumulative returns calculation.

-- cumsum (Cumulative Sum): If you use cumsum, you're calculating the cumulative sum of daily returns. This represents the total return over time, where each day's return is added to the previous total. It's more suitable for scenarios where you want to measure the total profit or loss over time.

-- cumprod (Cumulative Product): On the other hand, if you use cumprod, you're calculating the cumulative product of daily returns. This represents the overall growth or decline in the investment value over time, assuming you reinvest your returns. It's more suitable for scenarios where you want to measure the compounded return over time. So, if you want to understand how an investment grows or declines over time with reinvestment of returns, cumprod is the more appropriate choice.

```python
# The code is calculating the cumulative returns for a list of companies and plotting the "Buy and Hold" graph for each company.

def calculate_cumulative_returns_and_plot(companies_close):
    # Define a list of company symbols
    company_symbols = ["AMZN",	"GOOGL", "META", "MSFT", "NFLX"]

    for company_symbol in company_symbols:
    
        # Create a new DataFrame for the specified company's data
        company_data = companies_close[company_symbol].copy().to_frame()
        
        # Calculate daily returns using log returns
        company_data['Daily_Return'] = np.log(company_data / company_data.shift(1))
        
        # Drop NaN values
        company_data.dropna(inplace=True)
        
        # Calculate cumulative returns using cumsum
        company_data['Cumulative_Return'] = company_data['Daily_Return'].cumsum().apply(np.exp)
        
        # Plot the "Buy and Hold" graph
        company_data['Cumulative_Return'].plot(figsize=(12, 6))
        plt.title(f"{company_symbol} Buy and Hold (cumsum)", fontsize=15)
        plt.xlabel("Date")
        plt.ylabel("Cumulative Return")
        plt.show()
```

```python
calculate_cumulative_returns_and_plot(close)
```

```python
def calculate_cumulative_prod_returns_and_plot(companies_close):
    # Define a list of company symbols
    company_symbols = ["AMZN",	"GOOGL", "META", "MSFT", "NFLX"]

    for company_symbol in company_symbols:
    
        # Create a new DataFrame for the specified company's data
        company_data = companies_close[company_symbol].copy().to_frame()
        
        # Calculate daily returns using log returns
        company_data['Daily_Return'] = np.log(company_data / company_data.shift(1))
        
        # Drop NaN values
        company_data.dropna(inplace=True)
        
        # Calculate cumulative returns
        company_data['Cumulative_Return'] = (1 + company_data['Daily_Return']).cumprod() - 1
        
        # Plot the "Buy and Hold" graph
        company_data['Cumulative_Return'].plot(figsize=(12, 6))
        plt.title(f"{company_symbol} Buy and Hold (cumprod)", fontsize=15)
        plt.xlabel("Date")
        plt.ylabel("Cumulative Return")
        plt.show()
```

```python
calculate_cumulative_prod_returns_and_plot(close)
```

## Drawdowns:
A drawdown is a measure of the peak-to-trough decline in the value of an investment.

```python


def calculate_drawdown_and_plot(companies_close):
    # Define a list of company symbols
    company_symbols = ["AMZN", "GOOGL", "META", "MSFT", "NFLX"]
    
    drawdown_data = {}

    for company_symbol in company_symbols:
        # Create a new DataFrame for the specified company's data
        company_data = companies_close[company_symbol].copy().to_frame()
        
        # Calculate daily returns using log returns
        company_data['Daily_Return'] = np.log(company_data / company_data.shift(1))
        
        # Drop NaN values
        company_data.dropna(inplace=True)
        
        # Calculate cumulative returns using cumsum
        company_data['Cumulative_Return'] = company_data['Daily_Return'].cumsum().apply(np.exp)
        
        # Calculate previous peaks
        previous_peaks = company_data['Cumulative_Return'].cummax()
        
        # Calculate drawdowns
        drawdowns = (previous_peaks - company_data['Cumulative_Return']) / previous_peaks
        
        # Store the results in a dictionary
        drawdown_data[company_symbol] = {
            'Cumulative_Return': company_data['Cumulative_Return'],
            'Cumulative_Max': previous_peaks,
            'Drawdown': drawdowns
        }
        
        # Plot cumulative returns and cumulative max
        plt.figure(figsize=(12, 6))
        plt.plot(company_data.index, company_data['Cumulative_Return'], label="Cumulative Return")
        plt.plot(company_data.index, previous_peaks, label="Cumulative Max")
        plt.title(f"{company_symbol} Cumulative Return and Cumulative Max", fontsize=15)
        plt.xlabel("Date")
        plt.ylabel("Value")
        plt.legend()
        plt.show()


# Example usage:
# Assuming you have a DataFrame 'companies_close' with the historical daily prices for the selected companies
# call the function like this:
# drawdown_data = calculate_drawdown_and_plot(companies_close)
```

```python
calculate_drawdown_and_plot(close)
```

## SMA Strategy

```python
def test_strategy(companies_close, SMA):
    # Define a list of company symbols
    company_symbols = ["AMZN", "GOOGL", "META", "MSFT", "NFLX"]
        
    for company_symbol in company_symbols:
        # Create a new DataFrame for the specified company's data
        company_data = companies_close[company_symbol].copy().to_frame()
        
        # Calculate daily returns using log returns
        company_data['Daily_Return'] = np.log(company_data / company_data.shift(1))
        
          
        # Calculate and add the 50-day Simple Moving Average (SMA50)
        company_data['SMA_Small'] = company_data[company_symbol].rolling(int(SMA[0])).mean()
        
        # Calculate and add the 200-day Simple Moving Average (SMA200)
        company_data['SMA_Long'] = company_data[company_symbol].rolling(int(SMA[1])).mean()

        # Drop NaN values
        company_data.dropna(inplace=True)
        
        
        company_data["Position"] = np.where(company_data["SMA_Small"] > company_data["SMA_Long"], 1, -1)
        
        company_data["Strategy"] = company_data["Daily_Return"] * company_data["Position"].shift(1)
        
        company_data.dropna(inplace=True)
        
        ret = np.exp(company_data["Strategy"].sum())
        
        std = company_data["Strategy"].std() * np.sqrt(260)
        
        
        print(f"{company_symbol} Return is:", ret, "and Standard Deviation is:", std)
```

```python
test_strategy(close, (50,200))
```

## Strategy Adjusted with Long Bias
